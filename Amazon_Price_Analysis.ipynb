```python
import pandas as pd
import os

folder_path = "Downloads/Amazon_Data"

all_data = []

for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)
        
        df = pd.read_csv(file_path)
        
        # Add category column
        df['category'] = file.replace(".csv", "")
        
        all_data.append(df)

# Combine all files
final_df = pd.concat(all_data, ignore_index=True)

final_df.shape
```

```python
final_df.columns
```

```python
final_df.drop(['image', 'link', 'Unnamed: 0'], axis=1, inplace=True)
```

```python
final_df.info()
```

```python
final_df['discount_price'] = final_df['discount_price'].replace('[₹,]', '', regex=True)
final_df['discount_price'] = pd.to_numeric(final_df['discount_price'], errors='coerce')

final_df['actual_price'] = final_df['actual_price'].replace('[₹,]', '', regex=True)
final_df['actual_price'] = pd.to_numeric(final_df['actual_price'], errors='coerce')
```

```python
final_df['ratings'] = final_df['ratings'].str.extract('(\d+\.\d+)')
final_df['ratings'] = pd.to_numeric(final_df['ratings'], errors='coerce')
```

```python
final_df['no_of_ratings'] = final_df['no_of_ratings'].str.replace(',', '', regex=True)
final_df['no_of_ratings'] = final_df['no_of_ratings'].str.extract('(\d+)')
final_df['no_of_ratings'] = pd.to_numeric(final_df['no_of_ratings'], errors='coerce')
```

```python
final_df['ratings'] = final_df['ratings'].fillna(final_df['ratings'].mean())
```

```python
final_df['discount_price'] = final_df['discount_price'].fillna(final_df['actual_price'])
final_df['actual_price'] = final_df['actual_price'].fillna(final_df['discount_price'])
```

```python
final_df = final_df.dropna(subset=['discount_price', 'actual_price'], how='all')
```

```python
final_df['no_of_ratings'] = final_df['no_of_ratings'].fillna(0)
```

```python
final_df['discount_percent'] = (
    (final_df['actual_price'] - final_df['discount_price']) / final_df['actual_price']
) * 100
```

```python
final_df['discount_percent'] = final_df['discount_percent'].fillna(0)
```
